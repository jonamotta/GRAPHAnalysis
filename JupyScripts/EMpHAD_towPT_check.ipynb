```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
indir = '/home/llr/cms/motta/HGCAL/CMSSW_11_1_0/src/GRAPHAnalysis/L1BDT/hdf5dataframes/matched'

FE = 'threshold'

inFileTraining_dict = {
    'threshold'    : indir+'/GluGluHHTo2b2Tau_PU200_th_towers.hdf5',
    'mixed'        : indir+'/'
}
```

```python
store_tr = pd.HDFStore(inFileTraining_dict[FE], mode='r')
dfTraining = store_tr[FE]
store_tr.close()
```

```python
plt.figure(figsize=(8,8))
plt.hist(dfTraining['tower_pt'], bins=np.arange(0,20,1), color='limegreen', histtype='step', lw=2, alpha=0.5)
plt.hist(dfTraining['tower_etEm']+dfTraining['tower_etHad'], bins=np.arange(0,20,1), color='fuchsia', histtype='step', lw=2, alpha=0.5)
plt.grid(linestyle=':')
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.hist(dfTraining['tower_pt']-dfTraining['tower_etEm']-dfTraining['tower_etHad'], color='limegreen', histtype='step', lw=2, alpha=1)
plt.grid(linestyle=':')
plt.show()
```
